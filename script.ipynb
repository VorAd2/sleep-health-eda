```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
df[['Systolic_BP','Diastolic_BP']] = df['Blood Pressure'].str.split('/', expand= True)
df['Systolic_BP'] = df['Systolic_BP'].astype(float)
df['Diastolic_BP'] = df['Diastolic_BP'].astype(float)
print(df[['Systolic_BP', 'Diastolic_BP']].dtypes)

df.head()
```

```python
df['BMI Category'] = df['BMI Category'].replace('Normal Weight', 'Normal')
print(df['BMI Category'].unique())
```

```python
def qualitative(column):
    print(f"--- Variable: {column} ---")
    abs_freq = df[column].value_counts()
    rel_freq = df[column].value_counts(normalize=True) * 100 
    table = pd.DataFrame({'Frequência Absoluta': abs_freq, 'Frequência Relativa(%)': rel_freq})
    print(table)
    print('\n')
```

```python
qualitative('Gender')
qualitative('Occupation')
qualitative('BMI Category')
qualitative('Sleep Disorder')
```

```python
plt.figure(figsize=(10, 6))

order = ['Normal', 'Overweight', 'Obese']
sns.countplot(data=df, x='BMI Category', order=order, palette='magma')
plt.title('Distribuição de IMC')
plt.xlabel('Categoria')
plt.ylabel('Qtd. pessoas')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

order = df['Occupation'].value_counts().index
sns.countplot(data=df, y='Occupation', order=order, palette='mako')
plt.title('Distribuição de Profissões (Ordem Decrescente)')
plt.xlabel('Quantidade de Pessoas')
plt.ylabel('Profissão')
plt.show()
```
